# Comptage des data existantes pour chaque année
Somme à partir de tous les fichiers d'un dossier - 
Puis option filtrage import / export

```python
%load_ext autoreload
%autoreload 2 
import matplotlib.pyplot as plt
import os
import numpy as np
import pandas as pd
from pprint import pprint
from projetressourcesagricoles.data import * 
from projetressourcesagricoles.params import * 
```

```python
liste_years = ['y' + str(y) for y in range(1961, 2021)]
```

```python
dosTrav = FAO_FOLDER
years_dict_cce = {}
numFich = 0
for nomFich in os.listdir(dosTrav):
    if not nomFich.startswith('.') and nomFich.endswith('.csv'):
        numFich += 1
        print(numFich, nomFich)
        chFich = os.path.join(dosTrav, nomFich)
        df = pd.read_csv(chFich, low_memory=False)

        for col in liste_years:
            nb_vals = df[col].count()
            nb_vals_tot = nb_vals + years_dict_cce.get(col, 0)
            years_dict_cce[col] = nb_vals_tot 
        
print('**************** Terminé ')
years_dict_cce
```

```python
# Même fonction avec filtrage export qté
years_dict_expqt = {}
numFich = 0
for nomFich in os.listdir(dosTrav):
    if not nomFich.startswith('.') and nomFich.endswith('.csv'):
        numFich += 1
        print(numFich, nomFich)
        chFich = os.path.join(dosTrav, nomFich)
        df = pd.read_csv(chFich, low_memory=False)

        for col in liste_years:
            nb_vals = df[df['Élément'] == 'Exportations - Quantité'][col].count()
            nb_vals_tot = nb_vals + years_dict_expqt.get(col, 0)
            years_dict_expqt[col] = nb_vals_tot 
        
print('**************** Terminé ')
years_dict_expqt
```

```python
dosTrav = PROD_FOLDER
years_dict_prod = {}
numFich = 0
for nomFich in os.listdir(dosTrav):
    if not nomFich.startswith('.') and nomFich.endswith('.csv'):
        numFich += 1
        print(numFich, nomFich)
        chFich = os.path.join(dosTrav, nomFich)
        df = pd.read_csv(chFich, low_memory=False)

        for col in liste_years:
            nb_vals = df[col].count()
            nb_vals_tot = nb_vals + years_dict_prod.get(col, 0)
            years_dict_prod[col] = nb_vals_tot 
        
print('**************** Terminé ')
years_dict_prod
```

```python
years_dict_cce_df = pd.DataFrame.from_dict(years_dict_cce, orient='index')
```

```python
years_dict_prod_df = pd.DataFrame.from_dict(years_dict_prod, orient='index')
```
