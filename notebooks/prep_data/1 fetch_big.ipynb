```python
# Récupération de la ligne d'en-tête du dataset des transits
headerslist = pd.read_csv('../../raw_data/FAO/data_origine/Commerce_MatricesCommerceDetaillees_F_Toutes_les_Données.csv',
                         encoding=ENCODING, 
                         nrows=1
                        ).columns
headerslist = list(headerslist)
# Mise en minuscules des titres d'années pour cohérence avec Production
yearslist = [year.lower() for year in headerslist[9:]]
headerslist = headerslist[:9]
headerslist += yearslist
headerslist
```

```python
# Génération des csv par code pays à partir des splits du dataset des transits
dosTrav = "/home/mickh/code/mathieudimartino/projet-ressources-agricoles/raw_data/FAO/big_split"
numFich = 0
for nomFich in os.listdir(dosTrav):
    numFich += 1
    print(numFich, nomFich)
    chFich = os.path.join(dosTrav, nomFich)
    df = pd.read_csv(chFich, encoding=ENCODING, names=headerslist, low_memory=False)
    liste_codes_decl = df['Code pays déclarant'].unique()
    for code in liste_codes_decl:
        print("********", code)
        df_tmp = df[df['Code pays déclarant']==code]
        nomFichOut = str(code) + '.csv'
        chFichOut = os.path.join(dosTrav, nomFichOut)
        df_tmp.to_csv(chFichOut, mode='a', index=False)
print('*********** Terminé ! ************')
```

```python
# Génération des csv par code pays à partir des productions par continents
dosTrav = "/home/mickh/code/mathieudimartino/projet-ressources-agricoles/raw_data/agriculture-crop-production/dosTrav"
for nomFich in os.listdir(dosTrav):
    print(nomFich)
    chFich = os.path.join(dosTrav, nomFich)
    df = pd.read_csv(chFich, encoding=ENCODING, low_memory=False)
    liste_codes_decl = df['area_code'].unique()
    for code in liste_codes_decl:
        print("********", code)
        df_tmp = df[df['area_code']==code]
        nomFichOut = str(code) + '.csv'
        chFichOut = os.path.join(dosTrav, nomFichOut)
        df_tmp.to_csv(chFichOut, mode='a', index=False)
print('*********** Terminé ! ************')
```
